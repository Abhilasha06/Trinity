```python
#web app works
from flask import Flask
from flask import request
from flask import render_template
import os
import face_recognition
import cv2

app = Flask(__name__)

@app.route('/')
def my_form():
    return render_template("my_form.html")

@app.route('/', methods=['POST'])
def my_form_post():
    st=[]
    data1 = request.form['file']
    imagepath = "C://Users//Abhilasha//"+data1



    img = cv2.imread(imagepath,0)
#gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
    faces = faceCascade.detectMultiScale(
        img,
        scaleFactor=1.3,
        minNeighbors=3,
        minSize=(30, 30)
    )

    print("[INFO] Found {0} Faces!".format(len(faces)))

    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

    status = cv2.imwrite('C:\\Users\\Abhilasha\\faces_detected.jpg', img)
    print("[INFO] Image faces_detected.jpg written to filesystem: ", status)


    images = os.listdir('E://aaa')

   
    

# load your image
    for (x, y, w, h)  in faces:
    
        p=cv2.imread(imagepath)
        frame = p[y:y+h,x:x+w]

        cv2.imwrite("C://Users//Abhilasha//st.jpg", frame)
        image_to_be_matched = face_recognition.load_image_file("C://Users//Abhilasha//st.jpg")
    
        encodings = face_recognition.face_encodings(image_to_be_matched)
        if len(encodings) > 0:
            image_to_be_matched_encoded = encodings[0]
        else:
            continue

# encoded the loaded image into a feature vector
   
    

# iterate over each image
        for image in images:
    # load the image
            current_image = face_recognition.load_image_file("E://aaa//" + image)
    # encode the loaded image into a feature vector
            current_image_encoded = face_recognition.face_encodings(current_image)[0]
    # match your image with the image and check if it matches
            result = face_recognition.compare_faces(
                [image_to_be_matched_encoded], current_image_encoded)
    # check if it was a match
            if result[0] == True:
                print("Matched: "+image)
                st.append(image)
    
        #else:
            #print("No match: "+image)
    list_1_string = ' '.join(st)
    return render_template('index.html', message=list_1_string)

   




if __name__ == '__main__':
    app.run()
```

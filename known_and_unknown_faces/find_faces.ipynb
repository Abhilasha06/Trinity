```python
import os
import face_recognition
import cv2
import matplotlib.pyplot as plt
```

```python
#reading the input image
imagepath = "C:\\Users\\Abhilasha\\Desktop\\cse.jpg"
img = cv2.imread(imagepath,0)
#cv2.imshow("abc",img)
#haar cascade face detector
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
faces = faceCascade.detectMultiScale(img, scaleFactor=1.3, minNeighbors=3, minSize=(30, 30))
```

```python
print("Found {0} Faces!".format(len(faces)))
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)
status = cv2.imwrite('C:\\Users\\Abhilasha\\Desktop\\faces_detected.jpg', img)
print("Image faces_detected.jpg written to filesystem: ", status)
```

```python
images = os.listdir('E://my_faces')
st=[]
for (x, y, w, h)  in faces:
    
    p=cv2.imread(imagepath,0)
    frame = p[y:y+h,x:x+w]
    plt.imshow(frame)
    plt.show()
    cv2.imwrite("C://Users//Abhilasha//Desktop//st.jpg", frame)
    image_to_be_matched = face_recognition.load_image_file("C://Users//Abhilasha//Desktop//st.jpg")
    #creating facial encodings for all the detected faces
    encodings = face_recognition.face_encodings(image_to_be_matched, num_jitters=100)
    print(encodings)
    if len(encodings) > 0:
         image_to_be_matched_encoded = encodings[0]
    else:
         continue
    for image in images:
        current_image = face_recognition.load_image_file("E://my_faces//" + image)
        # encode the loaded image into a feature vector
        current_image_encoded = face_recognition.face_encodings(current_image)[0]
        result = face_recognition.compare_faces([image_to_be_matched_encoded], current_image_encoded, tolerance = 0.5)
        if result[0] == True:
            print("Matched: "+image)
            st.append(image.split('.')[0])
    
list_1_string = ' '.join(st)
print("Found faces "+list_1_string)
```
